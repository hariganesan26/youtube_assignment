# getting the value from category dataset

```python
category_df.head()
```

```python
len(category_df)
```

```python
category_df["items"][0]
```

```python
items_list=category_df["items"].tolist()
```

```python
#Extracting the required information from the category dataframe
id_list=[]
title=[]
for item in items_list:
    id_list.append(int(item["id"]))
    title.append(item["snippet"]["title"])
```

```python
#Storing the information in a dataframe
new_category_df=pd.DataFrame({"category_title":title,"category_id":id_list})
```

```python
new_category_df.head(10)
```

```python
new_category_df.info()
```

# Now merge the category column to the dataset  on ( "category_id")

```python
most_ca_df=pd.merge(most_ca,new_category_df,on="category_id",how="inner")
most_ca_df.tail()
```

```python
least_ca_df=pd.merge(least_ca,new_category_df,on="category_id",how="inner")
least_ca_df.head()
```

```python
most_us_df=pd.merge(top_us,new_category_df,on="category_id",how="inner")
most_us_df.tail()
```

```python
least_us_df=pd.merge(least_us,new_category_df,on="category_id",how="inner")
least_us_df.head()
```

```python
most_ca_df.to_csv("most_pop_ca.csv")
```

```python
least_ca_df.to_csv("least_pop_ca.csv")
```

```python
most_us_df.to_csv("most_pop_us.csv")
```
